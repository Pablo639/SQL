# Data Types in SQL

If the data we add in the database (as INT, VARCHAR, etc) we have to be sure is the correct data in order to process it effectively.

### INT

An integer value

### NUMERIC

A precise number with fixed decimal points. 
- precision is the total number of digits
- scale is the number of digits to the right fo the decimal point

### TEXT

- A string of characters

### VARCHAR

- A variable-length (where n is the maximun length): 'Hello' with VARCHAR(10)
- provides a maximun character limit, but can store strings of any lenth up to that limit

### BOOLEAN

- True, False

### Date
- YYYY-MM-DD

### TIMESTAMP
- YYYY-MM-DD HH:MI:SS

### TIMESTAMP WITH TIME ZONE
- Similar to TIMESTAMP, but includes time zone information
