# Recent Sales >= 2020

```python
%%sql

SELECT
  orderdate,
  quantity * netprice * exchangerate AS net_revenue
FROM sales
WHERE
  orderdate::date >= '2020-01-01' -- to confirm the variable date actually has date type
LIMIT 5;
```

# Add Customer Info

```python
%%sql
-- We want to know who ordered
SELECT
  s.orderdate,
  s.quantity * s.netprice * s.exchangerate AS net_revenue,
  c.givenname,
  c.surname,
  c.countryfull,
  c.continent
FROM
  sales s
LEFT JOIN customer as c ON s.customerkey = c.customerkey
WHERE
  orderdate::date >= '2020-01-01'
LIMIT 5;
```

```python
%%sql
-- We want to see the product, too
SELECT
  s.orderdate,
  s.quantity * s.netprice * s.exchangerate AS net_revenue,
  c.givenname,
  c.surname,
  c.countryfull,
  c.continent,
  p.productkey,
  p.productname,
  p.categoryname,
  p.subcategoryname
FROM
  sales s
LEFT JOIN customer as c ON s.customerkey = c.customerkey
LEFT JOIN product as p ON s.productkey = p.productkey
WHERE
  orderdate::date >= '2020-01-01'
LIMIT 3;
```

# Which are high value clients?

```python
%%sql
-- We want to see the product, too
SELECT
  s.orderdate,
  s.quantity * s.netprice * s.exchangerate AS net_revenue,
  c.givenname,
  c.surname,
  c.countryfull,
  c.continent,
  p.productkey,
  p.productname,
  p.categoryname,
  p.subcategoryname,
  CASE
    WHEN s.quantity * s.netprice * s.exchangerate  > 10000 THEN 'HIGH'
    ELSE 'LOW'
  END AS client_value
FROM
  sales s
LEFT JOIN customer as c ON s.customerkey = c.customerkey
LEFT JOIN product as p ON s.productkey = p.productkey
WHERE
  orderdate::date >= '2020-01-01'
ORDER BY
  net_revenue DESC;
```
